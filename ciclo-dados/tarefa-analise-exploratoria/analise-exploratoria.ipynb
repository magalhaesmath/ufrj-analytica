```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Limpeza e tratamento dos dados

```python
# Importando os DataFrames:
bus_data = pd.read_csv("bus_data.csv", sep="|")
company_data = pd.read_csv("company_data.csv", sep=";")

# Preenchendo valores nulos em 'driver_name' e 'bus_route':
bus_data.fillna({'driver_name': "motorista não atribuído", 'bus_route': "rota não atribuída"}, inplace=True)

# Padronizando datas, ignorando erros:
bus_data['bus_built_date'] = pd.to_datetime(bus_data['bus_built_date'], errors='coerce')

# Fazendo o merge com 'bus_line_id' e 'id':
df = pd.merge(bus_data, company_data, left_on='bus_line_id', right_on='id', how='inner')

# Convertendo 'bus_integrity' para numérico e tratando valores inválidos:
df['bus_integrity'] = pd.to_numeric(df['bus_integrity'], errors='coerce')
df['bus_integrity'] = df['bus_integrity'].fillna(df['bus_integrity'].mean()) 

# Renomeando a coluna 'bus_price(R$)' para evitar problemas:
df.rename(columns={'bus_price(R$)': 'bus_price'}, inplace=True)

# Substituindo ',' por '.' e convertendo 'bus_price' para float:
df['bus_price'] = df['bus_price'].astype(str).str.replace(',', '.')
df['bus_price'] = pd.to_numeric(df['bus_price'], errors='coerce')

# Convertendo colunas para os tipos indicados:
df = df.convert_dtypes()

# Verificando os novos tipos de dados das colunas e a quantidade de nulos:
df.info()
```

```python
# Um panorama mais aprofundado, mas ainda generalizado, sobre os dados do DF:
df.describe()
```

```python
# A matriz de correlação das variáveis dos dados:
df.corr()
```

```python
# Vendo as 5 primeiras linhas do DF para sempre poder consultar a estrutura dele:
df.head()
```

### Relação entre **bus_price** e **bus_integrity**

```python
# Quero investigar se há alguma relação entre o preço do veículo e a sua integridade, ou seja, será que ônibus mais caros estão melhor conservados?
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='bus_price', y='bus_integrity', alpha=0.5)
plt.title('Relação entre o Preço do Ônibus e sua Integridade')
plt.show()

print("Correlação: \n", df[['bus_price', 'bus_integrity']].corr())
```

Percebemos, pela célula acima, que os dados estão espalhados de forma completamente aleatória, ou seja, **não há** relação linear clara entre as variáveis **bus_price** e **bus_integrity**.

### Relação entre **bus_price** e **bus_capacity**

```python
# Agora quero investigar se há relação entre o preço de um ônibus e sua capacidade de passageiros, ou seja, ônibus mais caros comportam mais passageiros?
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='bus_price', y='bus_capacity', alpha=0.5)
plt.title('Relação entre o Preço do Ônibus e sua Capacidade')
plt.show()

print("Correlação: \n", df[['bus_price', 'bus_capacity']].corr())
```

Percebemos, pela célula acima, que, mais uma vez, os dados estão espalhados de forma completamente aleatória, ou seja, **não há** relação linear clara entre as variáveis **bus_price** e **bus_capacity**.

### Relação entre **bus_price** e **bus_built_date**

```python
# Quero investigar se há relação entre o preço do ônibus e a data de sua fabricação: será que os ônibus desvalorizam com o passar do tempo?
# Primeiramente, preciso calcular a idade dos ônibus. Farei uma nova coluna, chamada bus_age, com a idade dos ônibus até 2025.
ano_atual = 2025
df['bus_age'] = ano_atual - df['bus_built_date'].dt.year
```

```python
# Verificando a correlação entre bus_price e bus_age:
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='bus_age', y='bus_price', alpha=0.5)
plt.title('Relação entre o Preço do Ônibus e sua Idade')
plt.show()

print("Correlação: \n", df[['bus_age', 'bus_price']].corr())
```

Dados sem relação linear explícita.

### Relação entre **bus_integrity** e **bus_start_time**
Será que há alguma relação entre a integridade dos ônibus e o turno de suas respectivas rotas? Ou seja: ônibus do período noturno têm integridade menor que os ônibus do período diurno? Ou o contrário?

```python
# Convertendo a coluna 'bus_start_time' para hora e criando uma nova coluna 'hora' para armazenar a hora de início:
df['hora_inicio'] = pd.to_datetime(df['bus_start_time'], format='%I:%M %p').dt.hour
```

```python
# Filtrando o df para comparar as integridades das linhas noturnas e diurnas:
df_diurno = df[((df['hora_inicio'] >= 6) & (df['hora_inicio'] < 18))]
df_noturno = df[((df['hora_inicio'] >= 18) | (df['hora_inicio'] < 6))]
```

```python
# Vou calcular a média de integridade nos turnos:
print(f"A média de integridade dos ônibus nas rotas diurnas é {df_diurno['bus_integrity'].mean()}")
print(f"A média de integridade dos ônibus nas rotas noturnas é {df_noturno['bus_integrity'].mean()}")
```

As médias, por si só, não dizem nada, pois variam muito pouco entre si. Vou tentar agora fazer um histograma da distribuição das integridades pelos horários.

```python
# Histograma das integridades - diurno:
df_diurno['bus_integrity'].hist()
plt.show()
```

```python
# Histograma das integridades - noturno:
df_noturno['bus_integrity'].hist()
plt.show()
```

Perceba como a distribuição é bem parecida em ambos os turnos. Sendo assim, sabemos que os turnos **não são** marcados por determinada integridade.
